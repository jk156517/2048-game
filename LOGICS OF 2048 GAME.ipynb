```python
import random
def start_game():
    mat=[]

    for i in range(4):
        mat.append([0]*4)

    return mat

def add_new_2(mat):

    r=random.randint()
    c=random.randint()

    while(mat[r][c]!=0):#if there is no empty space , we have to generate row and column number again
        r = random.randint()
        c = random.randint()

    mat[r][c]=2

def current_state_of_game(mat):

    #THIS IS GOING TO CHECK WHETHER THE GAME HAS BEEN WON BY THE USER IF ANY ELEMENT IN MATRIX BECOMES 2048
    for i in range(4):
        for j in range(4):
            if(mat[i][j]==2048):
                return 'WON'

    #THIS IS CHECKING WHETHER THERE IS EMPTY SPACE OR NOT , IF EMPTY SPACE THEN GAME HAS NOT BEEN OVER YET
    for i in range(4):
        for j in range(4):
            if(mat[i][j]==0):
                return 'GAME NOT OVER YET'


    #THIS IS CHECKING TILL RANGE 3 AS IF I IN 4 AND J IN 4 THEN THE NEXT INDEX GOES OUT OF RANGE,THEREFORE WE ARE
    #CHECKING TILL 3 FOR BOTH AND IF ADJACENT ELEMENTS ARE SAME THEN THE GAME IS NOT OVER YET
    for i in range(3):
        for j in range(3):
            if(mat[i][j]==mat[i][j+1] or mat[i][j]==mat[i+1][j]):
                return 'GAME NOT OVER YET'
    #THIS LOOP IS CHECKING FOR THE LAST COLUMN, KEEPING THE ROW NUMBER CONSTANT
    for j in range(4):
        if(mat[3][j]==mat[3][j+1]):
            return 'GAME NOT OVER YET'

    # THIS LOOP IS CHECKING FOR THE LAST ROW, KEEPING THE Column NUMBER CONSTANT
    for i in range(4):
        if(mat[i][3]==mat[i+1][3]):
            return 'GAME NOT OVER YET'

    #IF ANY OF THE LOOP ISN'T SATISFYING THAT MEANS WE HAVE LOST THE GAME . HENCE,
    return 'LOST'

#THIS FUNCTION IS FOR PUTTING ALL THE ZEROS (EMPTY SPACES TO END) AND NON-ZERO ELEMENTS IN STARTING
def compress(mat):

    changed=False
    new_mat=[]
    for i in range(4):
        new_mat.append([0]*4)

    for i in range(4):
        pos=0 #THIS VARIABLE IS KEEPING THE COUNT OF NON-ZERO VALUE

        for j in range(4):
            if(mat[i][j]!=0):
                new_mat[i][pos]=mat[i][j]
                pos=pos+1
            if j!=pos:
                changed=True

    return new_mat,changed

#THIS FUNCTION IS GOING TO MERGE THE ELEMENTS WHICH ARE SAME i.e., MULTIPLYING THE VALUE BY 2 AND PUTTING THE ZERO VALUE
def merge(mat):

    changed=False
    for i in range(4):
        for j in range(3):
            if(mat[i][j]==mat[i][j+1] and mat[i][j]!=0):
                mat[i][j]=mat[i][j]*2
                mat[i][j+1]=0
                changed=True

    return mat,changed

#THIS FUNCTION IS TAKING THE TRANSPOSE(row-> column and column->rows) OF THE MATRIX
def transpose(mat):

    new_mat=[]
    for i in range(4):
        new_mat.append([])

        for j in range(4):
            new_mat.append([j][i])

    return new_mat

#THIS FUNCTION IS TAKING THE REVERSE(row numbers if 2 4 8 2 to 2 8 4 2)OF THE MATRIX

def reverse(mat):

    new_mat=[]

    for i in range(4):
        new_mat.append([])
        for j in range(3):
            new_mat[i].append(mat[i][4-j-1])

    return new_mat

#for left move, we have to compress, then merge and then compress again
def move_left(grid):
    new_grid,changed1=compress(grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2
    new_grid=compress(new_grid)
    return new_grid,changed

#FOR RIGHT MOVE, WE WILL FIRST REVERSE THE MATRIX AND THEN APPLY THE MOVES OF LEFT CASE AND THEN REVERSE AGAIN
def move_right(grid):
    reversed_grid=reverse(grid)
    new_grid, changed1 = compress(grid)
    new_grid, changed2 = merge(new_grid)
    changed = changed1 or changed2
    final_grid=reverse(new_grid)
    return final_grid,changed

#FOR UP MOVE , WE WILL HAVE TO TAKE TRANSPOSE , THEN COMPRESS,MERGE AND THEN COMPRESS AGAIN
def move_up(grid):
    transposed_grid=transpose(grid)
    new_grid,changed1=compress(transposed_grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2
    new_grid=compress(new_grid)
    final_transposed_grid=transpose(new_grid)
    return final_transposed_grid,changed

#FOR DOWN MOVE, WE WILL HAVE TO FIRST TRANSPOSE ,THEN REVERSE, THEN COMPRESS, MERGE ADN THEN COMPRESS AGAIN AND THEN REVERSE
#AND THEN TAKE THE TRANSPOSE
def move_down(grid):
    transposed_grid=transpose(grid)
    reversed_grid=reverse(grid)
    new_grid,changed1=compress(reversed_grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2
    new_grid=compress(new_grid)
    final_reversed_grid=reverse(new_grid)
    final_transpoed_grid=transpose(final_reversed_grid)
    return final_transpoed_grid,changed



size=400
grid_len=4
grid_padding=10

background_color_game="red"
background_color_cell_empty="yellow"

background_color_dict={2: "red", 4: 'blue', 8: 'pink',16: "green" , 32: "red", 64: "purple", 128: "blue", 256: "green"
                       ,512: "orange" , 1024: "blue", 2048: "cyan"}

cell_color_dict={2: "purple" , 4: 'red', 8: 'yellow',16: "red" , 32: "green", 64: "black", 128: "brown", 256: "pink"
                       ,512: "orange" , 1024: "blue", 2048: "red"}

font=("VERDANA", 40, "BOLD")

KEY_UP="'w'"
KEY_DOWN="'s'"
KEY_LEFT="'a'"
KEY_RIGHT="'d'"

from tkinter import Frame,Label,CENTER



class game2048(Frame):
    def __init__(self):
        Frame.__init__(self)

        self.grid()
        self.master_title('2048')
        self.master.bind("<KEY>",self.key_down)

        self.commands={c.KEY_UP:LOGICSOF2048.move_up, c.KEY_DOWN: LOGICSOF2048.move_down, c.KEY_LEFT : LOGICSOF2048.moveleft
                       ,c.KEY_RIGHT: LOGICSOF2048.move_right}

        self.grid_cells=[]
        self.init_grid()
        self.init_matrix()
        self.update_grid_cells()

    def init_grid(self):
        background=Frame(self,bg=c.background_color_game,width=c.size,height=c.size)

        background.grid()

        for i in range(c.grid_len):
           grid_row=[ ]
           for j in row(c.grid_len):
               cell=Frame(background,bg=c.background_color_cell_empty,width=c.size/c.grid_len,height=c.size/c.grid_len)

               cell.grid(row=i,column=j,padx=c.grid_padding,pady=c.grid_padding)

               t=Label(master=cell,text=" ",bg=c.background_color_cell_empty,justify=CENTER,font=c.FONT,width=5,height=2)

               t.grid()
               grid_row.append(t)

           self.grid_cells.append(grid_row)

    def init_matrix(self):
        self.matrix=LOGICSOF2048.start_game()
        LOGICSOF2048.add_new_2(self.matrix)
        LOGICSOF2048.add_new_2(self.matrix)

    def update_grid_cells(self):
        for i in range(c.grid_len):
            for j in range(c.grid_len):
                new_number=self.matrix[i][j]
                if new_number==0:
                    self.grid_cells[i][j].configure(text="",bg=c.background_color_cell_empty)
                else:
                    self.grid_cells[i][j].configure(text=str(new_number),bg=c.background_color_dict[new_number],fg=c.
                                                    cell_color_dict[new_number])

        self.update_idletasks()

    def key_down(self,event):
        key= repr(event.char)

        changed=True

        if key in self.commands:
            self.matrix,changed=self.commands[repr(event.char)](self.matrix)

        if changed:
            LOGICSOF2048.add_new_2(self.matrix)
            self.update_grid_cells()
            changed=False

        if LOGICSOF2048.get_current_state(self.matrix)=='WON':
            self.grid_cells[1][1].configure(text="you" , bg=c.background_color_cell_empty)

            self.grid_cells[1][2].configure(text="won!" , bg=c.background_color_cell_empty)

        if LOGICSOF2048.get_current_state(self.matrix)=='LOST':
            self.grid_cells[1][1].configure(text="you" , bg=c.background_color_cell_empty)
            self.grid_cells[1][2].configure(text="lost!", bg=c.background_color_cell_empty)


gamegrid=game2048()






```
